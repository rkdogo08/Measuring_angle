```
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```
image=cv2.imread("/content/q.jpg")
```

# **Gaussian filter, Bilateral filter comparison**
가우시안 필터를 이용하여 노이즈의 제거 시에 상세 정보가 손상되어 검출되는 에지 위치의 정확성이 떨어진다.


```
gau_img=cv2.GaussianBlur(image,(19,19),0)

a=plt.subplots(1,2,figsize=(20, 20))
a[1][0].axis("off")
a[1][0].imshow(cv2.cvtColor(gau_img,cv2.COLOR_BGR2RGB))

bilateral_img=cv2.bilateralFilter(image,20,40,40)
a[1][1].axis("off")
a[1][1].imshow(cv2.cvtColor(bilateral_img,cv2.COLOR_BGR2RGB))


```

(좌) Gaussian filter (우) Bilateral filter

# **Bilateral filter**
양방향 필터는 에지를 보존하면서 노이즈를 제거하는 비선형 필터이다. 

양방향 필터도 가우시안 필터와 마찬거지로 일정 크기의 필터에 가중치를 부여하여 이웃하는 픽셀들을 고려한다. 차이점은 이웃하는 픽셀과의 색 강도, 깊이거리를 고려하여 가중치를 부여한다는 점이다.

void bilateralFilter( InputArray src, int d, double sigmaColor, double sigmaSpace, int borderType = BORDER_DEFAULT )

- d : 필터링을 수행 할 필터의 크기.
- sigmaColor : 색공간의 크기. 필터크기 내의 범위에서 더 멀리있는 색까지 혼합되어 넓은 영역이 반균등한 색을 가지게 되는 것을 의미한다.
- sigmaSpace : 거리공간의 크기. 파라미터의 값이 클수록 가까이 있는 픽셀뿐만 아니라 더 멀리 있는 픽셀의 영향을 받는 것을 의미한다.
 
 시그마값이 클수록 영향이 더 커지는 것을 의미한다.


```
a=plt.subplots(1,3,figsize=(30, 20))
a[1][0].axis("off")
a[1][0].imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))

bilateral_img=cv2.bilateralFilter(image,10,20,20)
a[1][1].axis("off")
a[1][1].imshow(cv2.cvtColor(bilateral_img,cv2.COLOR_BGR2RGB))

bilateral_img=cv2.bilateralFilter(image,20,40,40)
a[1][2].axis("off")
a[1][2].imshow(cv2.cvtColor(bilateral_img,cv2.COLOR_BGR2RGB))
```

(왼쪽부터) 원본이미지, (d=10, sigmaColor=20,sigmaSpace=10), (d=20, sigmaColor=40, sigmaSpace=40)
